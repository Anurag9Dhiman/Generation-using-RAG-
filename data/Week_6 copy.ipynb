Each bad_*.py file shows a violation of one SOLID principle.



### 1. SRP (bad_srp.py)

Class: Order

Problem: Handles 3 jobs - total calculation, database save, email

Violation: Single Responsibility Principle


### 2. OCP (bad_ocp.py)

Class: DiscountCalculator

Problem: New discounts need editing method with if/elif

Violation: Open/Closed Principle


### 3. LSP (bad_lsp.py)

Classes: Rectangle, Square

Problem: Square overrides width and height which breaks substitutability

Violation: Liskov Substitution Principle


### 4. ISP (bad_isp.py)

Interface: Worker (work, eat)

Problem: Robot forced to implement eat()

Violation: Interface Segregation Principle


### 5. DIP (bad_dip.py)

Classes: PasswordReminder, MySQLConnection

Problem: High-level class depends directly on low-level class

Violation: Dependency Inversion Principle


### Conclusion

SRP: too many roles in one class

OCP: requires code modification

LSP: broken inheritance

ISP: fat interface

DIP: no abstraction

Part 4: Reflective Questions

1. Which SOLID principle was hardest to apply and why?

The hardest principle to apply was the Dependency Inversion Principle (DIP).
It was challenging to identify all high-level modules that depended directly on low-level concrete classes
and then refactor them to depend on abstractions or interfaces.
Ensuring proper abstractions while keeping the design simple required careful planning.

2. What UML relationship was tricky to model?

The tricky UML relationship was the inheritance between Shape, Rectangle, and Square while respecting Liskov Substitution Principle (LSP).
We had to make sure that Square could be used anywhere a Rectangle or Shape is expected without breaking behavior.
This required abstracting a base class (Shape) and carefully modeling implementations in the diagram.

3. How did TDD improve/refine your design?

TDD helped refine the design by forcing us to think about **testable units** and clear responsibilities.
Writing tests first highlighted areas where classes were doing too much, leading to SRP refactoring.
It also ensured that interfaces and dependencies were designed properly, improving adherence to DIP and making the code more reliable.
